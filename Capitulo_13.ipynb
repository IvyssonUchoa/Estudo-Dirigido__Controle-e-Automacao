## 13.2 Sistemas com dados amostrados

<p>Um <strong>sistema de dados amostrados</strong> é aquele em que parte de sua operação ocorre com dados discretos, resultantes da amostragem de sinais contínuos. Esses sistemas são fundamentais no controle digital, onde computadores são utilizados para executar algoritmos de controle.</p>

<p>Nesses sistemas, a amostragem ocorre em intervalos de tempo fixos <em>T</em>, denominados <strong>período de amostragem</strong>. Assim, o sinal de entrada contínuo <em>r(t)</em> é amostrado e convertido em uma sequência de valores discretos <em>r(kT)</em>, em que <em>k</em> representa um número inteiro.</p>

<p>Um exemplo típico de sistema de controle digital envolve três componentes principais:</p>

<ul>
  <li>Um <strong>conversor analógico-digital (A/D)</strong>, que transforma o sinal contínuo em um sinal discreto;</li>
  <li>Um <strong>computador digital</strong>, responsável por processar os dados;</li>
  <li>Um <strong>conversor digital-analógico (D/A)</strong>, que reconverte o sinal processado para sua forma contínua.</li>
</ul>


## 13.3 A Transformada Z

<p>A transformada z é uma ferramenta matemática utilizada para analisar sistemas de controle digital que operam com dados amostrados. Como os computadores digitais recebem dados apenas em intervalos de tempo específicos, é necessário um método para descrever e analisar o desempenho desses sistemas.

<p>A transformada de Laplace de um sinal amostrado, r∗(t), é dada por:

$$
\mathcal{E}\{r^*(t)\} = \sum_{k=0}^{\infty} r(kT) e^{-ksT}
$$

<p> A transformada z é definida pela relação 

$$
z = e^{sT}
$$

<p>que mapeia o plano s para o plano z. Assim, a transformada z de uma função f(t) é expressa como:

$$
\mathcal{Z}\{f(t)\} = F(z) = \sum_{k=0}^{\infty} f(kT) z^{-k}
$$

<p>Esta transformada possui propriedades semelhantes às da transformada de Laplace e pode ser usada para analisar a estabilidade e a resposta transitória de um sistema. Uma tabela de transformadas z e suas propriedades está disponível para facilitar o cálculo

<p><strong>Transformada Z de Funções Exponenciais e Senoidais</strong>

<p>Vamos determinar a transformada z de uma função exponencial decrescente 

$e^{-at}$ :

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Parâmetros da função exponencial
a = 1
T = 1
z = np.linspace(0.1, 2, 400)

# Transformada z
F_z = 1 / (z - np.exp(-a * T))

# Plotar a transformada z
plt.figure()
plt.plot(z, np.abs(F_z))
plt.title('Transformada z de e^{-at}')
plt.xlabel('z')
plt.ylabel('|F(z)|')
plt.grid()
plt.show()
```

<p>Transformada Z de uma Função Senoidal Vamos determinar a transformada z de uma função seno sin(wt):

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

omega = 1

# Transformada z
F_z = (z * np.sin(omega * T)) / (z**2 - 2*z*np.cos(omega * T) + 1)

# Plotar a transformada z
plt.figure()
plt.plot(z, np.abs(F_z))
plt.title('Transformada z de \sin(\omega t)')
plt.xlabel('z')
plt.ylabel('|F(z)|')
plt.grid()
plt.show()
```

## 13.4 Análise de estabilidade no plano Z

<p>Um sistema de controle contínuo e linear é estável se todos os polos da sua função de transferência de malha fechada, T(s), estiverem no semiplano esquerdo do plano s. A relação entre o plano s e o plano z é dada pela transformação 

$z = e^{sT} = e^{(\sigma + j\omega)T}$. 

<p>Isso implica que 

$|z| = e^{\sigma T} \quad \text{e} \quad \underline{LZ} = \omega T$


<p>Com base nessa relação, o semiplano esquerdo do plano s, onde σ<0, corresponde à região interna do círculo unitário no plano z, onde a magnitude de z está entre 0 e 1. O eixo imaginário do plano s corresponde ao círculo unitário no plano z

<p>Dessa forma, a regra de estabilidade para sistemas amostrados é que um sistema é estável se todos os polos da sua função de transferência de malha fechada, 
T(z), estiverem localizados no interior do círculo unitário do plano z. Para determinar a estabilidade, a equação característica do sistema, q(z)=1+G(z)=0, deve ser analisada para encontrar as raízes (polos). Por exemplo, para um sistema de segunda ordem com coeficientes reais, as condições necessárias e suficientes para ter todas as raízes no interior do círculo unitário são:

∣q(0)∣<1, q(1)>0 e q(−1)>0


## 13.5 Desempenho de um sistema de segunda ordem com dados amostrados

<p>Considere-se o desempenho de um sistema de segunda ordem amostrado com um extrapolador de ordem zero, quando o processo a controlar é:

$$
G_p(s) = \frac{K}{s(\tau s + 1)}
$$

<p>Obtém-se, então, G(z) para um valor não especificado do período de amostragem T:

$$
G(z) = \frac{K \{(z - E)[T - \tau(z - 1)] + \tau(z - 1)^2\}}{(z - 1)(z - E)}
$$

<p>onde 

$$E = e^{-T / T}$$


<p>A estabilidade do sistema é analisada considerando a equação característica:

$$q(z) = z^2 + z\{K[T - T(1 - E)] - (1 + E)\} + K[T(1 - E) - T E] + E = 0$$


<p>Como o o polinômio q(z)
 é do segundo grau e tem todos os coeficientes reais, as condições necessárias e suficientes para ter todas as raízes no inteiror do círculo unitário são

∣q(0)∣<1, q(1)>0 e q(−1)>0

## 13.6 Sistemas a malha fechada com compensação digital por computador

<p>Um sistema amostrado em malha fechada, que utiliza um computador digital para aprimorar seu desempenho, possui a seguinte função de transferência:

$\frac{Y(z)}{R(z)} = T(z) = \frac{G(z)D(z)}{1 + G(z)D(z)}$

<p>A função de transferência do computador é representada por:

$\frac{U(z)}{E(z)} = D(z)$


<p><strong>Exemplo: Projeto de Sistema Amostrado com Compensador Digital</strong>

<p> O processo é representado pela função de transferência abaixo, considerando a retenção de ordem zero e T = 1:

$G_p(s) = \frac{1}{s(s + 1)}$

<p>Será considerado o controlador PID com função de transferência de domínio s

$$\frac{U(s)}{X(s)} = G_c(s) = K_1 + \frac{K_2}{s} + K_3s.$$

<p>Podemos empregar uma aproximação discreta tanto para a derivada quanto para a integral, a fim de viabilizar a implementação digital desse controlador.

<p>Para calcular a derivada em relação ao tempo, utilizamos a regra da diferença atrasada:

$$u(kT)=\frac{dx}{dt}|_{t=kT} \approx \frac{1}{T}[x(kT)-x(k-1)T)].$$

$$U(z)=\frac{(1-z^{-1})}{T}X(z)=\frac{(z-1)}{Tz}X(z).$$

```python
import numpy as np
import control as ctl
import matplotlib.pyplot as plt

# Definindo a função de transferência
num = [0.3678 + 0.2644]  # Corrigido
den = [1 - 1 + 0.6322]  # Corrigido

sys = ctl.TransferFunction(num, den)

# Definindo o intervalo de tempo para a resposta
T = np.arange(0, 8, 0.1)

# Calculando a resposta ao degrau
yout, T_out = ctl.step_response(sys, T)

# Plotando a resposta
plt.plot(T_out, yout)
plt.title('Resposta ao Degrau')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python

import numpy as np
import control as ctl
import matplotlib.pyplot as plt
from control import pade, minreal

# Definindo a função de transferência do sistema contínuo
numg = np.array([1])
deng = np.array([1, 1, 0])

# Aproximação de Pade de ordem 2 para a função de transferência
nd, dd = pade(1, 2)

# Criando a função de transferência discretizada
numd = np.array(dd) - np.array(nd)
dend = np.convolve(np.array([1, 0]), dd)

# Aplicando a minimização de ordem (minreal)
function1 = minreal(ctl.TransferFunction(numd, dend))

# Criando o sistema em série com a função de transferência 'function1'
function2 = ctl.series(function1, ctl.TransferFunction(numg, deng))

# Adicionando a retroalimentação
lastfunction = ctl.feedback(function2)

# Obtendo a resposta ao degrau
y, t = ctl.step_response(lastfunction)

# Plotando o gráfico
plt.grid()
plt.xlabel("Tempo")
plt.ylabel("y(t)")
plt.axis([0, 20, 0, 1.5])
plt.plot(t, y) 
plt.legend(title='')
plt.show()
```
