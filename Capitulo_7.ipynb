```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Defini√ß√£o do numerador e denominador da fun√ß√£o de transfer√™ncia G(s)H(s)
num = [0.5, 1]  # (1/2)s + 1
den = [0.25, 1, 0]  # s(1/4s + 1) = (1/4)s¬≤ + s


# Criando a fun√ß√£o de transfer√™ncia GH(s)
GH = ctrl.TransferFunction(num, den)


# Plotando o Lugar das Ra√≠zes
plt.figure(figsize=(8,6))
ctrl.rlocus(GH, grid=True)
plt.title("Lugar das Ra√≠zes do Sistema de Controle")
plt.xlabel("Parte Real")
plt.ylabel("Parte Imagin√°ria")
plt.show()
```

## 7.3 Projeto de Par√¢metro pelo M√©todo do Lugar das Ra√≠zes

<p>O <strong>Projeto de Par√¢metros pelo M√©todo do Lugar das Ra√≠zes</strong> √© uma extens√£o da t√©cnica tradicional do lugar das ra√≠zes, que normalmente analisa a varia√ß√£o de um √∫nico par√¢metro (geralmente o ganho <em>K</em>). Esse m√©todo possibilita investigar o efeito de m√∫ltiplos par√¢metros, como <em>Œ±</em> e <em>Œ≤</em>, na posi√ß√£o das ra√≠zes da equa√ß√£o caracter√≠stica de um sistema din√¢mico.</p>

<p>A ideia principal √© adaptar a equa√ß√£o caracter√≠stica para isolar o par√¢metro de interesse, transformando-a na forma padr√£o do lugar das ra√≠zes, <em>1 + F(s) = 0</em>. Caso um par√¢metro n√£o apare√ßa isoladamente como um coeficiente, a equa√ß√£o pode ser rearranjada para que ele se torne um fator multiplicativo.</p>


<p>Para lidar com m√∫ltiplos par√¢metros, a abordagem √© realizada de forma sequencial. Por exemplo, considerando dois par√¢metros, <em>Œ±</em> e <em>Œ≤</em>:</p>

<ol>
  <li>Inicialmente, um dos par√¢metros (por exemplo, <em>Œ≤</em>) √© fixado em um valor inicial, geralmente zero, e o lugar das ra√≠zes √© tra√ßado em fun√ß√£o do outro par√¢metro (por exemplo, <em>Œ±</em>).</li>
  <li>Com base nos requisitos de desempenho, √© selecionado um valor adequado para <em>Œ±</em>, e as ra√≠zes resultantes desse primeiro lugar das ra√≠zes passam a ser os polos para a etapa seguinte.</li>
  <li>Em seguida, o lugar das ra√≠zes √© tra√ßado em fun√ß√£o do segundo par√¢metro (<em>Œ≤</em>), utilizando os polos determinados na etapa anterior. Um valor apropriado para <em>Œ≤</em> √© ent√£o escolhido.</li>
</ol>

<p>Este m√©todo permite projetar sistemas com m√∫ltiplos par√¢metros ajust√°veis, como √© o caso do controlador PID, que possui tr√™s par√¢metros ajust√°veis (KP‚Äã, KI‚Äã, KD‚Äã)

## 7.4 Sensibilidade e lugar das ra√≠zes

<p>As varia√ß√µes nos par√¢metros de um sistema influenciam a posi√ß√£o das ra√≠zes, e medir essas varia√ß√µes √© fundamental, pois as ra√≠zes da equa√ß√£o caracter√≠stica representam os modos dominantes da resposta transit√≥ria do sistema.

<p>A sensibilidade de uma raiz em rela√ß√£o a um par√¢metro no sistema
ùëá
(
ùë†
) pode ser definida como:

$S_{K}^{r_i} = \frac{\partial r_i}{\partial \ln K} = \frac{\partial r_i}{\partial K / K}$

<p>O c√°lculo da sensibilidade das ra√≠zes em um sistema de controle pode ser realizado de forma direta utilizando os m√©todos do lugar das ra√≠zes.

<p>Neste exemplo, ser√° considerado um sistema de segunda ordem representado por uma fun√ß√£o de transfer√™ncia simples, no qual o ganho
ùêæ ser√° ajustado para projetar o comportamento desejado.

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

num = [1]  # K
den = [1, 2, 1]  # s^2 + 2s + 1


# Criando a fun√ß√£o de transfer√™ncia
G_H = ctrl.TransferFunction(num, den)


# Plotando o Lugar das Ra√≠zes para diferentes valores de K
plt.figure(figsize=(8,6))


# Plotando o Lugar das Ra√≠zes com uma grade para facilitar a visualiza√ß√£o
ctrl.rlocus(G_H, grid=True)


plt.title("Lugar das Ra√≠zes do Sistema com Ganho K Ajustado")
plt.xlabel("Parte Real")
plt.ylabel("Parte Imagin√°ria")
plt.show()
```

## 7.5 Controladores de Tr√™s Termos (PID)

<p>O controlador PID √© largamente empregado em sistemas de controle autom√°tico. Sua atua√ß√£o baseia-se em tr√™s termos fundamentais ‚Äî proporcional, integral e derivativo ‚Äî que, combinados, permitem ajustar a sa√≠da do sistema de forma eficiente. A fun√ß√£o de transfer√™ncia do controlador PID √© expressa por:

$C(s) = K_p + \frac{K_i}{s} + K_d s$

<p>Onde (K_p), (K_i), e (K_d) s√£o os ganhos proporcional, integral e derivativo, respectivamente.

<p>O lugar das ra√≠zes pode ser utilizado para ajustar esses par√¢metros, garantindo que o sistema atenda aos requisitos de estabilidade e desempenho.

## 7.6 Exemplo de Projeto: Sistema de Controle de um Manipulador de Laser

<p>
  Considere o sistema de controle de um <strong>manipulador a laser</strong>, o qual deve ser ajustado para executar movimentos com rapidez e alta precis√£o.
  O projeto do controlador pode ser realizado por meio do <strong>m√©todo do lugar das ra√≠zes</strong>, ajustando-se o ganho K e os par√¢metros do controlador PID de forma a posicionar as ra√≠zes adequadamente no plano s,
  de modo a satisfazer os crit√©rios de desempenho, como tempo de subida, ultrapassagem e estabilidade.
</p>

<p>
  A seguir, ser√° realizada a <strong>simula√ß√£o</strong> da resposta do sistema a uma entrada em rampa, com o objetivo de analisar a ultrapassagem e o tempo de assentamento.
</p>


```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Definindo os par√¢metros do sistema
K = 10  # Ganho
num = [K]  # Numerador da fun√ß√£o de transfer√™ncia
den = [1, 15, 50, 50*K]  # Denominador da fun√ß√£o de transfer√™ncia (com K = 10)


# Criando a fun√ß√£o de transfer√™ncia do sistema
system = ctrl.TransferFunction(num, den)


# Definindo a entrada em rampa: r(t) = A * t (com A = 1 mm/s)
A = 1  # Amplitude da entrada em rampa
time = np.linspace(0, 10, 500)  # Tempo de simula√ß√£o
rampa = A * time  # Entrada em rampa: r(t) = A * t


# Resposta do sistema a uma entrada em rampa
time, response = ctrl.forced_response(system, time, rampa)


# Plotando a resposta
plt.figure(figsize=(10,6))
plt.plot(time, response, label='Resposta do Sistema')
plt.title('Resposta do Sistema a uma Entrada em Rampa (K = 10)')
plt.xlabel('Tempo [s]')
plt.ylabel('Posi√ß√£o [mm]')
plt.grid(True)
plt.legend()
plt.show()
```
